# 1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

print("Nguyễn Hữu Toàn – 6351071071")
df = pd.read_excel("tb_tracking.xlsx")
print(df.head())
```

```python
print("Nguyễn Hữu Toàn – 6351071071")
print(df.info())
```

```python
print("Nguyễn Hữu Toàn – 6351071071")
print(df.describe())
```

# 2

```python
total = df.isnull().sum().sort_values(ascending=False)
percent = round(df.isnull().sum()/df.isnull().count()*100, 1).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', '%'])
print("Nguyễn Hữu Toàn – 6351071071")
print(missing_data.head(5))
```

# 3

```python
df.columns = ["id","device_id","co_level","humidity","temperature","time"]
df["time"] = pd.to_datetime(df["time"], dayfirst=True, errors="coerce")
df["minutes"] = df["time"].dt.hour*60 + df["time"].dt.minute

plt.figure()
plt.scatter(df["minutes"], df["co_level"], s=8, alpha=0.6)
plt.xlabel("minutes")
plt.ylabel("CO level")
plt.title("CO vs minutes")
plt.grid(True)
plt.tight_layout()
plt.show()
print("Nguyễn Hữu Toàn – 6351071071")

plt.figure()
plt.scatter(df["minutes"], df["humidity"], s=8, alpha=0.6)
plt.xlabel("minutes")
plt.ylabel("Humidity")
plt.title("Humidity vs minutes")
plt.grid(True)
plt.tight_layout()
plt.show()
print("Nguyễn Hữu Toàn – 6351071071")

plt.figure()
plt.scatter(df["minutes"], df["temperature"], s=8, alpha=0.6)
plt.xlabel("minutes")
plt.ylabel("Temperature")
plt.title("Temperature vs minutes")
plt.grid(True)
plt.tight_layout()
plt.show()
print("Nguyễn Hữu Toàn – 6351071071")
```

# 4

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='device_id', y='co_level', data=df)
plt.title("CO level theo thiết bị")
plt.xlabel("Device ID")
plt.ylabel("CO level")
plt.show()
print("Nguyễn Hữu Toàn – 6351071071")
```

# 5

```python
stats_device = df.groupby('device_id')[['co_level','humidity','temperature']].describe()
print("Nguyễn Hữu Toàn – 6351071071")
print(stats_device)
```
